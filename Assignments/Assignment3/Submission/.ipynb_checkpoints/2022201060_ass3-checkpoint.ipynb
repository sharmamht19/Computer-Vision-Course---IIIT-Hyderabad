```python
def download_video(video_url, output_dir):
    yt = YouTube(video_url)
    video = yt.streams.filter(file_extension='mp4', resolution='480p').first()
    video_path = video.download(output_dir)

    # Check if the video was downloaded successfully
    if os.path.exists(video_path):
        new_file_path = os.path.join(output_dir, "input_video.mp4")
        os.rename(video_path, new_file_path)
        print("Video downloaded successfully.")
        return new_file_path
    else:
        print("Error downloading video.")
        return None
def extract_frames(video_path, output_dir, duration=30):
    os.makedirs(output_dir, exist_ok=True)
    # Extract frames from the video
    ffmpeg_command = f"ffmpeg -i {video_path} -vf fps=24/1 -t {duration} {output_dir}/frame_%03d.png"
    result = os.system(ffmpeg_command)

    if result == 0:
        print("Frames extracted successfully.")
        return True
    else:
        print("Error extracting frames.")
        return False
```

```python

def detect_faces_and_save_video(frames_folder, output_folder, output_video_name):
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')


    frame_path = f'{frames_folder}/frame_001.png'
    first_frame = cv2.imread(frame_path)
    height, width, _ = first_frame.shape

    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    output_video_path = os.path.join(output_folder, output_video_name)
    out = cv2.VideoWriter(output_video_path, fourcc, 24.0, (width, height))

    total_processing_time = 0
    for i in range(1, 721):

        frame = cv2.imread(f'{frames_folder}/frame_{i:03d}.png')

        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        start_time = time.time()
        faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.3, minNeighbors=8, minSize=(15, 15))
        end_time = time.time()

        processing_time = end_time - start_time
        total_processing_time += processing_time

        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        out.write(frame)
        print(f"Frame {i}: Processing time = {processing_time:.2f} seconds")

    out.release()
#     cv2.destroyAllWindows()

    print(f"Total processing time for all frames: {total_processing_time:.2f} seconds")
```

## 1.4 - Association-based tracking.
Tracking can be used to associate face detections across time and understand that it is the same character appearing across multiple frames of the movie. We will explore a simple way to perform tracking.

### 1.4.1 Generate face tracks by comparing face detections in two consecutive frames and associating them based on IoU scores.
You may want to associate faces only when IoU > 0.5. Do consider what happens when there are multiple face detections in both frames. Start new tracks for faces not seen in the previous frame. End existing tracks when faces are not visible in the next frame. How many unique tracks did you create in the first 30 seconds?

```python
def detect_faces(frame):
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.3, minNeighbors=8, minSize=(15, 15))
    return faces

def compute_iou(box1, box2):
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[0] + box1[2], box2[0] + box2[2])
    y2 = min(box1[1] + box1[3], box2[1] + box2[3])
    intersection_area = max(0, x2 - x1 + 1) * max(0, y2 - y1 + 1)
    box1_area = (box1[2] + 1) * (box1[3] + 1)
    box2_area = (box2[2] + 1) * (box2[3] + 1)
    iou = intersection_area / float(box1_area + box2_area - intersection_area)
    return iou

def associate_face_tracks(frames_folder):
    tracks = {}
    output_folder = "output"
    output_video_name = "tracked_output.mp4"
    output_video_path = os.path.join(output_folder, output_video_name)

    # Get information about the first frame
    frame_path = os.path.join(frames_folder, "frame_001.png")
    first_frame = cv2.imread(frame_path)
    height, width, _ = first_frame.shape

    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_video_path, fourcc, 24.0, (width, height))

    track_id = 1  # Initialize track id
    unique_tracks_created = 0  # Counter for unique tracks

    for i in range(1, len(os.listdir(frames_folder))):  # Loop through frames
        frame_path = os.path.join(frames_folder, f"frame_{i:03d}.png")
        frame = cv2.imread(frame_path)

        curr_faces = detect_faces(frame)

        # Get faces in the next frame
        next_frame_path = os.path.join(frames_folder, f"frame_{(i+1):03d}.png")
        next_frame = cv2.imread(next_frame_path)
        next_faces = detect_faces(next_frame)
        # Store matching tracks for current frame
        matched_tracks = {}
        # Associate faces with tracks
        for face1 in next_faces :
            for face2 in curr_faces:
                if compute_iou(face1, face2) > 0.5:
                    if tuple(face2) in tracks.values():
                        for key, value in tracks.items():
                            if value == tuple(face2):
                                tracks[key] = tuple(face1)
                                matched_tracks[tuple(face2)] = key
                    else:
                        matched_tracks[tuple(face2)] = track_id
                        tracks[track_id] = tuple(face1)
                        unique_tracks_created += 1
                        track_id += 1

        # Draw faces and write track IDs on the frame
        for (x, y, w, h) in curr_faces:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
            cv2.putText(frame, str(matched_tracks.get(tuple((x, y, w, h)), '')), (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        out.write(frame)

    out.release()
    print("Tracked video saved successfully!")
    print("Number of unique tracks created in the first 30 seconds:", unique_tracks_created)
```

```python
frames_folder = "input_frames"
associate_face_tracks(frames_folder)
```

Study the YOLO object detector in detail. Understand and explain (in your own words, no plagiarism!) how single-shot detectors such as YOLO are different from the R-CNN series that we learned in class. There are more than 8 versions of the YOLO series. Pick any 3 and understand and explain (in your words) the differences between them.

### YOLO (You Only Look Once)

YOLO, short for "You Only Look Once," is an object detection algorithm widely used in computer vision tasks. Unlike traditional methods like R-CNN series, which examine various parts of an image multiple times, YOLO takes a different approach by analyzing the entire image once and directly predicting bounding boxes and class probabilities.

#### How YOLO Works:

1. **Divide and Predict:** YOLO splits the image into a grid of cells and predicts bounding boxes and class probabilities for each grid cell. Each cell takes responsibility for detecting objects within its boundaries.

2. **Single Pass:** YOLO processes the entire image in one pass, making it exceptionally fast compared to methods that require multiple passes through the image.

3. **Bounding Box Prediction:** For each grid cell, YOLO predicts multiple bounding boxes along with confidence scores for each box, indicating the likelihood of containing an object.

4. **Class Prediction:** YOLO also predicts class probabilities for each bounding box, specifying the type of object it believes is present in the box.

5. **Non-Max Suppression:** YOLO employs non-max suppression after predictions to eliminate redundant bounding boxes, retaining only the most confident ones.

### Difference between Single-shot Detectors (e.g., YOLO) and R-CNN Series:

Single-shot detectors like YOLO process the entire image at once to predict bounding boxes and class probabilities. They skip a separate region proposal step, making them faster but potentially less accurate for small objects and localization.

R-CNN series, on the other hand, follows a two-stage approach. It first generates region proposals and then extracts features and classifies them. While more accurate, especially for small objects, it's slower due to the two-stage process.

### Comparison of YOLO Versions:

1. **Architecture:**
   - **YOLOv1:** YOLOv1 was revolutionary for combining tasks like drawing bounding boxes and identifying class labels into a single, end-to-end network.
   - **YOLOv2:** YOLOv2 stepped up with improvements like BatchNorm, better resolution, and anchor boxes, boosting its performance.
   - **YOLOv3:** Building on its predecessors, YOLOv3 added objectivity estimation to predict bounding boxes, along with connections to network layers and predictions at different detail levels.

2. **Speed and Accuracy:**
   YOLO models excel in both speed and accuracy.
   - **YOLOv1:** Hit a remarkable 45 frames per second (FPS) prediction speed on a Titan X GPU with an average accuracy of 63.4 mAP, outshining other real-time detectors.
   - **YOLOv2:** Upgraded to 76.8 mAP on the VOC 2007 dataset at 67 FPS with 416×416 input data resolution, surpassing its predecessor.
   - **YOLOv3:** Despite a slightly lower mAP of 28.2, YOLOv3 maintained a fast 45 FPS on a Titan X GPU with 320×320 input data resolution, offering comparable accuracy to others but with faster processing.

3. **Techniques:**
   - **YOLOv1:** Marked the beginning, integrating object detection tasks.
   - **YOLOv2:** Utilized BatchNorm, improved resolution, and anchor boxes for better performance.
   - **YOLOv3:** Introduced techniques like objectivity estimation and multi-level predictions, especially useful for small object detection.

```python

def create_yaml_file(dataset_root, train_images_dir, train_labels_dir, val_images_dir, val_labels_dir, classes, download_url=None, file_name='config.yaml'):
    config = {
        'path': dataset_root,
        'train': train_images_dir,
        'val': val_images_dir,
        'nc': len(classes),
        'names': {i: class_name for i, class_name in enumerate(classes)}
    }

    yaml_file_path = os.path.join(dataset_root, file_name)
    with open(yaml_file_path, 'w') as file:
        documents = yaml.dump(config, file)

class_names = [
    'duck'
]
```

```python
def analyze_yolo_models(yaml_file, epochs=10):

    # Step 1: Creating a YOLOv8n Model
    model_v8n = YOLO('yolov8n')  # Create a YOLOv8n model

    # Step 2: Train the YOLOv8n Model
    results_v8n = model_v8n.train(data=yaml_file, epochs=epochs, imgsz=640)

    # Step 3: Get the number of parameters and convolutional layers of the YOLOv8n model
    params_v8n = sum(p.numel() for p in model_v8n.parameters())
    conv_layers_v8n = sum(1 for m in model_v8n.modules() if isinstance(m, nn.Conv2d))

    # Step 4: Load the YOLOv8m Model
    model_v8m = YOLO('yolov8m')  # Load the YOLOv8m model

    # Step 5: Get the number of parameters and convolutional layers of the YOLOv8m model
    params_v8m = sum(p.numel() for p in model_v8m.parameters())
    conv_layers_v8m = sum(1 for m in model_v8m.modules() if isinstance(m, nn.Conv2d))

    # Return the results
    return {
        "yolov8n_model": model_v8n,
        "yolov8n_parameters": params_v8n,
        "yolov8n_conv_layers": conv_layers_v8n,
        "yolov8m_model": model_v8m,
        "yolov8m_parameters": params_v8m,
        "yolov8m_conv_layers": conv_layers_v8m,
        "yolov8n_results": results_v8n
    }
```

```python
def prepare_training_dataset(extracted_dir,root_dir):
    # Check if the extracted directory exists
    if not os.path.exists(extracted_dir):
        print("Error: Dataset directory not found.")
        return

    # Define writable directories for train1 and train2
#     root_dir = "/content/drive/MyDrive/datasets"
    train1_images_dir = root_dir + "/dataset1/images/train"
    train1_labels_dir = root_dir + "/dataset1/labels/train"
    train2_images_dir = root_dir + "/dataset2/images/train"
    train2_labels_dir = root_dir + "/dataset2/labels/train"
    train1_images_val_dir = root_dir + "/dataset1/images/val"
    train1_labels_val_dir = root_dir + "/dataset1/labels/val"
    train2_images_val_dir = root_dir + "/dataset2/images/val"
    train2_labels_val_dir = root_dir + "/dataset2/labels/val"

    # Create train1 directories
    os.makedirs(train1_images_dir, exist_ok=True)
    os.makedirs(train1_labels_dir, exist_ok=True)
    os.makedirs(train1_images_val_dir, exist_ok=True)
    os.makedirs(train1_labels_val_dir, exist_ok=True)

    # Define paths to the original training set
    original_train_images_dir = os.path.join(extracted_dir, "images", "train")
    original_train_labels_dir = os.path.join(extracted_dir, "labels", "train")
    original_train_val_images_dir = os.path.join(extracted_dir, "images", "val")
    original_train_val_labels_dir = os.path.join(extracted_dir, "labels", "val")

    # Check if the directories exist
    if not os.path.exists(original_train_images_dir) or not os.path.exists(original_train_labels_dir):
        print("Error: Original training directories not found.")
        return

    # Randomly select samples for train1 dataset
    train1_indices = random.sample(os.listdir(original_train_images_dir), 100)
    for image_file in train1_indices:
        # Construct full paths for the image and label files
        image_path = os.path.join(original_train_images_dir, image_file)
        label_file = os.path.splitext(image_file)[0] + ".txt"
        label_path = os.path.join(original_train_labels_dir, label_file)

        # Copy the image and label files to the train1 directories
        shutil.copy(image_path, train1_images_dir)
        shutil.copy(label_path, train1_labels_dir)

    # Check if train2 directories exist
    if not os.path.exists(train2_images_dir) or not os.path.exists(train2_labels_dir):
        # Create train2 directories
        os.makedirs(train2_images_dir, exist_ok=True)
        os.makedirs(train2_labels_dir, exist_ok=True)
        os.makedirs(train2_images_val_dir, exist_ok=True)
        os.makedirs(train2_labels_val_dir, exist_ok=True)

        # Randomly select 400 samples for train2 dataset
        train2_indices = random.sample(os.listdir(original_train_images_dir), 400)
        for image_file in train2_indices:
            # Construct full paths for the image and label files
            image_path = os.path.join(original_train_images_dir, image_file)
            label_file = os.path.splitext(image_file)[0] + ".txt"
            label_path = os.path.join(original_train_labels_dir, label_file)

            # Copy the image and label files to the train2 directories
            shutil.copy(image_path, train2_images_dir)
            shutil.copy(label_path, train2_labels_dir)
    # copy entire validation set to train1 and train2
    for image_file in os.listdir(original_train_val_images_dir):
        # Construct full paths for the image and label files
        image_path = os.path.join(original_train_val_images_dir, image_file)
        label_file = os.path.splitext(image_file)[0] + ".txt"
        label_path = os.path.join(original_train_val_labels_dir, label_file)

        # Copy the image and label files to the train1 directories
        shutil.copy(image_path, train1_images_val_dir)
        shutil.copy(label_path, train1_labels_val_dir)
        # Copy the image and label files to the train2 directories
        shutil.copy(image_path, train2_images_val_dir)
        shutil.copy(label_path, train2_labels_val_dir)
    # config.yaml file for train1
    create_yaml_file(root_dir + "/dataset1/", train1_images_dir, train1_labels_dir, train1_images_val_dir, train1_labels_val_dir, class_names, "")
    # config.yaml file for train2
    create_yaml_file(root_dir + "/dataset2/", train2_images_dir, train2_labels_dir, train2_images_val_dir, train2_labels_val_dir, class_names, "")


    print("Training datasets created successfully.")

    # Return the paths to the training datasets
    return train1_images_dir, train1_labels_dir, train2_images_dir, train2_labels_dir
```

```python
def create_training_yaml_file(dataset_root, train_images_dir, train_labels_dir, val_images_dir, val_labels_dir, classes, download_url=None, file_name = 'yolo_dataset.yaml'):
    config = {
        'path': dataset_root,
        'train': train_images_dir,
        'val': val_images_dir,
        'test': None,  # You can specify the path to the test set if available
        'names': {i: class_name for i, class_name in enumerate(classes)},
        'download': download_url
    }

    with open(file_name , 'w') as file:
        yaml.dump(config, file)
```

```python
def train_yolo_variants(extracted_dir, models, epochs=10, device=None):
    # Define paths
    train1_yaml = extracted_dir + "/dataset1/config.yaml"
    train2_yaml = extracted_dir + "/dataset2/config.yaml"

    # Initialize dictionary to store trained models
    trained_models = {}

    # Train on train1 dataset
    for dataset_name, yaml_file in [("train1", train1_yaml), ("train2", train2_yaml)]:
        trained_models[dataset_name] = {}
        print(f"\n{'='*20} Training on {dataset_name} dataset {'='*20}\n")


        for model_name, model_file in models.items():
            print(f"Training {model_name}:")
            model = YOLO(model_file)
            model.train(data=yaml_file, epochs=epochs)
            trained_models[dataset_name][model_name] = model

            print(f"\n{'*'*10} Model {model_name} trained successfully {'*'*10}\n")

    return trained_models
```

#### 2.4.3 - 6 “best” model checkpoints after the training.
Train all models for the same number of epochs (e.g., 20). Report and compare the results (AP50) across the 6 variants on both the training and validation set. (a) How does increasing dataset size impact performance? (b) Does the bigger model perform better? Why or why not?

```python
def compare_model_performance(evaluation_results):
    train_ap50s = {}
    val_ap50s = {}
    individual_train_ap50s = {}
    individual_val_ap50s = {}

    # Loop through the evaluation results and extract AP50 values
    for dataset_name, dataset_results in evaluation_results.items():
        train_ap50s[dataset_name] = []
        val_ap50s[dataset_name] = []
        individual_train_ap50s[dataset_name] = {}
        individual_val_ap50s[dataset_name] = {}

        for model_name, model_result in dataset_results.items():
            # Extract AP50 values for training and validation sets
            train_ap50 = model_result.box.map50
            val_ap50 = model_result.box.map50

            # Append AP50 values to respective lists
            train_ap50s[dataset_name].append(train_ap50)
            val_ap50s[dataset_name].append(val_ap50)

            # Store individual AP50 values for each model
            individual_train_ap50s[dataset_name][model_name] = train_ap50
            individual_val_ap50s[dataset_name][model_name] = val_ap50

    # Calculate average AP50 values for each model variant on both training and validation sets
    avg_train_ap50s = {model_name: sum(ap50_list) / len(ap50_list) for model_name, ap50_list in train_ap50s.items()}
    avg_val_ap50s = {model_name: sum(ap50_list) / len(ap50_list) for model_name, ap50_list in val_ap50s.items()}

    return avg_train_ap50s, avg_val_ap50s, individual_train_ap50s, individual_val_ap50s
```

```python
avg_train_ap50s, avg_val_ap50s, individual_train_ap50s, individual_val_ap50s = compare_model_performance(evaluation_results)

# Print average AP50 values for each model variant on both training and validation sets
print("Average AP50 values on the training set:")
for model_name, avg_ap50 in avg_train_ap50s.items():
    print(f"{model_name}: {avg_ap50}")

print("\nAverage AP50 values on the validation set:")
for model_name, avg_ap50 in avg_val_ap50s.items():
    print(f"{model_name}: {avg_ap50}")

# Print individual AP50 values for each model variant on both training and validation sets
print("\nIndividual AP50 values on the training set:")
for dataset_name, model_ap50s in individual_train_ap50s.items():
    print(f"\nDataset: {dataset_name}")
    for model_name, ap50 in model_ap50s.items():
        print(f"Model: {model_name}: {ap50}")

print("\nIndividual AP50 values on the validation set:")
for dataset_name, model_ap50s in individual_val_ap50s.items():
    print(f"\nDataset: {dataset_name}")
    for model_name, ap50 in model_ap50s.items():
        print(f"Model: {model_name}: {ap50}")
```

### 2.5 Impact of augmentations.
(i) Does the default run use augmentations? Which ones and what do they do?
- Hint: analyze the output train logs (watch out for albumentations, an augmentations library).


The default run for training the YOLOv8n model with pretrained weights includes various augmentation techniques to enhance the diversity and quality of the training data as per the training logs we can see above and below. Here are the specific augmentation parameters and their values:

1. **HSV (Hue, Saturation, Value) Adjustment**:
   - hsv_h: 0.015
   - hsv_s: 0.7
   - hsv_v: 0.4
   
   These parameters control the adjustments made to the hue, saturation, and value components of the image colors.

2. **Rotation**:
   - degrees: 0.0
   
   This parameter controls the degree of rotation applied to the images.

3. **Translation**:
   - translate: 0.1
   
   This parameter controls the extent of translation (shifting) applied to the images.

4. **Scaling**:
   - scale: 0.5
   
   This parameter controls the degree of scaling applied to the images.

5. **Shear Transformation**:
   - shear: 0.0
   
   This parameter controls the degree of shearing (distortion) applied to the images.

6. **Perspective Transformation**:
   - perspective: 0.0
   
   This parameter controls the degree of perspective transformation applied to the images.

7. **Flipping**:
   - fliplr: 0.5
   - flipud: 0.0
   
   These parameters control the probability of horizontal and vertical flipping of the images.

8. **Color Space Adjustment**:
   - bgr: 0.0
   
   This parameter controls the adjustment of the image colors in the BGR color space.

9. **Mosaic Effect**:
   - mosaic: 1.0
   
   This parameter controls the degree of applying the mosaic effect to the images.

10. **Mixing Images**:
   - mixup: 0.0
   
   This parameter controls the degree of mixing images during training.

11. **Copy-Pasting**:
   - copy_paste: 0.0
   
   This parameter controls the degree of copy-pasting objects within images during training.

12. **Automatic Augmentation**:
   - auto_augment: randaugment
   
   This parameter specifies the type of automatic augmentation technique used, which in this case is "randaugment."

13. **Erasing**:
   - erasing: 0.4
   
   This parameter controls the degree of random erasing applied to the images.

14. **Cropping**:
   - crop_fraction: 1.0
   
   This parameter controls the degree of cropping applied to the images.


```python
def train_with_augmentations(dataset, epochs=20):
    
    # Create the YOLO model
    model = YOLO('yolov8n.pt')
    
    # Train the model with augmentations
    model.train(data=dataset, epochs=epochs, pretrained=True)
    
    # Validate the trained model
    results = model.val()
    
    return model, results

def train_without_augmentations(dataset, epochs=20):

    # Create the YOLO model
    model = YOLO('yolov8n.pt')
    
    # Train the model without augmentations
    model.train(data=dataset, epochs=epochs, pretrained=True,
                hsv_h=0, hsv_s=0, hsv_v=0, translate=0, scale=0,
                fliplr=0, mosaic=0, erasing=0, crop_fraction=0 )
    
    # Validate the trained model
    results = model.val()
    
    return model, results
```

```python
# Train with augmentations
epochs = 10
yaml_file1 = "/kaggle/working/dataset1/config.yaml"
trained_model_with_aug, results_with_aug = train_with_augmentations(yaml_file1,epochs)

# Train without augmentations
trained_model_without_aug, results_without_aug = train_without_augmentations(yaml_file1,epochs)
```
